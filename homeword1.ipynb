vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần 
diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 
chữ số phần thập phân

bài 3:Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem trước cấu trúc lựa chọn: if….else, ma)

```python
c = input()
if 'A' <= c <= 'Z':
    print(c.lower())
elif 'a' <= c <= 'z':
    print(c.upper())
else:
    print("Không phải chữ cái.")
```

bài 4:Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái 
không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c} 
không phải là kí tự alphabet.

```python
c = input()
if c.isalpha():
    print(f"{c} là kí tự alphabet")
else:
    print(f"{c} không phải là kí tự alphabet")
```

bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương ứng của A. Xét TH khi c = ‘A’ là đặc biệt

```python
c = input()
if c == 'a':
    print("'A' là đặc biệt.")
else:
    print(chr(ord(c) + 31))
```

bài 6: Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”

```python
a = float(input())
b = float(input())
c = float(input())
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    s = (p * (p - a) * (p - b) * (p - c))** 0.5
    print(f"{s:.1f}")
else:
    print("Không phải 3 cạnh của tam giác")
```

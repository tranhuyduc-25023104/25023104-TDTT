tính tống các số nguyên từ 1 đến n. (n <= 1000)

bài 6:Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn 
[a, b]. Mỗi dòng là một bộ test (a b).

```python
a, b = map(int, input().split())
tong = 0
for i in range(a, b+1):
    if i < 2:
        continue

    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            break
    else:
        tong += i
            
print(tong)
```

bài 7:Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n

```python
n = int(input())
unt_max = None
for i in range(1,n + 1):
    if n % i == 0:
        for j in range(2, int(i**0.5) + 1):
            if i%j == 0:
                break
        else:
            if unt_max == None or unt_max < i:
                unt_max = i
print(unt_max)
```

bài 8: Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.

```python
def reverse(n):
    rev_n = 0
    while n > 0:
        rev_n = rev_n*10 + (n % 10)
        n //= 10
    return rev_n

def palindrome(n):
    return reverse(n) == n

n = int(input())
dem = 0

while not palindrome(n):
    n = n + reverse(n)
    dem += 1

print(dem, reverse(n))
```

bài 9: Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number).

```python
def ktra(n):
    while n > 0:
        so_cuoi = n % 10
        con_lai = n // 10
        
        while con_lai > 0:
            so_khac = con_lai % 10
            if so_cuoi == so_khac:
                return False
            con_lai = con_lai // 10
        
        n = n // 10
    return True

n = int(input())
co_so = False
da_in = False

i = 1
while i * i <= n:
    so_cp = i * i
    if ktra(so_cp):
        if da_in:
            print(" ", end="")
        print(so_cp, end="")
        co_so = True
        da_in = True
    i = i + 1

if not co_so:
    print("no number")
```

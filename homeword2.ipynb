Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là
độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh 
cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn
hình.

```python
a, b, c = map(int, input().split())
if a > 0 and b >0 and c > 0 and a + b > c and a + c > b and b + c > a:
    import math
    if a == b == c:
        loai = "tam giác đều"
    elif a == b or b == c or c == a:
        loai = "tam giác cân"
    elif a*a + b*b == c*c or b*b + c*c == a*a or a*a + c*c == b*b:
        loai = "tam giác vuông"
    else:
        loai = "tam giác thường"
    chu_vi = a + b + c
    p = chu_vi / 2
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"diện tích và chu vi {loai} là: {dien_tich:.2f} {chu_vi}")
else:
    print("khong la ba canh cua mot tam giac")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math
a, b, c = map(int, input().split())
delta = b*b - 4 * a * c
if delta > 0:
    x1 = (-b + math.sqrt(delta))
    x2 = (-b - math.sqrt(delta))
    print(f"pt có 2 nghiệm phân biệt x1 = {x1:.2f} và x2 = {x2:.2f}")
elif delta == 0:
    x = -b / 2
    print(f"pt có nghiệm kép là x = {x:.2f}")
else:
    print("pt vô nghiệm")
```

Bài 8: Giải hệ phương trình tuyến tính:
ax + by = m
cx + dy = n

```python
a, b, c, d, m, n = map(int, input().split())
D = a*d - b*c
Dx = m*d - n*b
Dy = a*n - c*m
if D == 0 and (Dx != 0 or Dy != 0):
    print("hpt vô nghiệm")
elif D == Dx == Dy == 0:
    print("hpt vô số nghiệm")
else:
    x = round(Dx / D, 2)
    y = round(Dy / D, 2)
    print("hpt có nghiệm (x, y) = :", (x, y))
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết 
ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao 
nhiêu giây.

```python
x = int(input())
h = x // 3600
p = (x - h * 3600) // 60
s = x - h * 3600 - p * 60
print(h, p, s)
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math
x0, y0 = map(int, input().split())
r = int(input())
x, y = map(int, input().split())
kc= math.sqrt((x - x0)*(x - x0) + (y - y0) * (y - y0))
if kc <= r:
    print(f"điểm A({x}, {y}) thuộc đường tròn")
else:
     print(f"điểm A({x}, {y}) không thuộc đường tròn")
```
